### Count Plot

## Profit Analysis

### Profit Analysis of Ship Mode

```python
ship_mode = pd.DataFrame(df.groupby(['Ship Mode'])['Profit'].sum()).reset_index()
ship_mode
```

```python
ship_mode.plot(x='Ship Mode', y='Profit', kind='bar')
```

### From the above plot we can infer that Standard Class has the highest number of profit and Same Day has the lowest number of profit

## Profit Analysis of Segment

```python
segment = pd.DataFrame(df.groupby(['Segment'])['Profit'].sum()).reset_index()
segment
```

```python
segment.plot(x='Segment', y='Profit', kind='bar')
```

### From the above plot we can infer that Consumer Segment has the highest number of profit and Home Office has the lowest number of profit

## Profit Analysis of Category

```python
category = pd.DataFrame(df.groupby(['Category'])['Profit'].sum()).reset_index()
category
```

```python
category.plot(x='Category', y='Profit', kind='bar')
```

### From the above plot we can infer that Technology has the highest number of profit and Furniture has the lowest number of profit

## Profit Analysis of Region

```python
region = pd.DataFrame(df.groupby(['Region'])['Profit'].sum()).reset_index()
region
```

```python
region.plot(x='Region', y='Profit', kind='bar')
```

### From the above plot we can infer that West Region has the highest number of profit and South Region has the lowest number of profit

## Total Profit

```python
total_profit = pd.DataFrame(df['Profit']).describe()
total_profit
```

## Top 10 Cities with Highest Profit

```python
max_profit = pd.DataFrame(df.groupby(['City'])['Profit'].sum()).reset_index().sort_values('Profit', ascending=False).head(10)
max_profit
```

## Top 10 Cities with Lowest Profit

```python
min_profit = pd.DataFrame(df.groupby(['City'])['Profit'].sum()).reset_index().sort_values('Profit', ascending=False).tail(10)
min_profit
```

## Analysis of Profit,Sales and Discounts

### Segmentwise analysis

```python
df['Segment'].value_counts()
```

```python
df_segment = df.groupby(['Segment'])[['Sales', 'Discount', 'Profit']].mean()
df_segment
```

```python
df_segment.plot.pie(subplots = True,
                   autopct = '%1.1f%%',
                   figsize=(20,18),
                   labels = df_segment.index)
plt.title('Segmentwise Analysis of Sales, Discounts, Profit', fontsize='20')
```
